```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plant1_gendf = pd.read_csv('Plant_1_Generation_Data.csv')
plant1_weatdf = pd.read_csv('Plant_1_Weather_Sensor_Data.csv')
plant2_gendf = pd.read_csv('Plant_2_Generation_Data.csv')
plant2_weatdf = pd.read_csv('Plant_2_Weather_Sensor_Data.csv')
```

```python
## Datetime formating
plant1_gendf['DATE_TIME'] = pd.to_datetime(plant1_gendf['DATE_TIME'], dayfirst=True)
plant1_weatdf['DATE_TIME'] = pd.to_datetime(plant1_weatdf['DATE_TIME'], dayfirst=True)
plant2_gendf['DATE_TIME'] = pd.to_datetime(plant2_gendf['DATE_TIME'], dayfirst=True)
plant2_weatdf['DATE_TIME'] = pd.to_datetime(plant2_weatdf['DATE_TIME'], dayfirst=True)
```

```python
from matplotlib.ticker import MaxNLocator
plt.figure(figsize=[15,8])
plt.plot(plant1_gendf['DATE_TIME'],plant1_gendf['AC_POWER'])
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```
